```python
raw, spans = read_article(111111111)
```

```python
token_list = raw.split()
```

```python
token_list
```

```python
first_letter_index = []
index = 0
for token in token_list:
    success = False
    while not success:
        if raw[index] != '\n':
            first_letter_index.append(index)
            index += len(token) + 1
            success = True
        else:
            index += 1
```

```python
for i in range(len(token_list)):
    print(token_list[i], first_letter_index[i], raw[first_letter_index[i]])
```

```python
binary_labels = [0] * len(token_list)
for i in range(len(first_letter_index)):
    for span in spans:
        if (span[0] <= first_letter_index[i] < span[1]):
            binary_labels[i] = 1
```

```python
labelled_tokens_list = []
for i in range(len(binary_labels)):
    labelled_tokens_list.append([token_list[i], binary_labels[i]])
```

```python
for i in range(len(labelled_tokens_list)-1):
    if (labelled_tokens_list[i][0][0] == '\"') & (labelled_tokens_list[i+1][1] == 1):
        labelled_tokens_list[i][1] = 1
```
